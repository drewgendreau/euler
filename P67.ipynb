```julia
a = Array(Vector{Int}, 0)
open("p067_triangle.txt") do f
    b = 1
    for i in eachline(f)
        z = split(i, " ")
        z = map(strip, z)
        z = map(parse, z)
        push!(a, z)
    end
end
```

```julia
a
```

```julia
for current in (length(a)-1):-1:1
    last = current+1
    for i in 1:length(a[current])
        a[current][i] += max(a[last][i], a[last][i+1])
    end
end
```
